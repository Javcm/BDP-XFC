### Carga de librerías para el análisis 

```python
import pandas as pd
import numpy as np
from datetime import datetime
import re
import matplotlib.pyplot as plt
```

### Cargar bases de datos que utilizaremos

```python
df_courses=pd.read_csv('data/courses.csv', dtype = {'ID': str,'CourseName': str}).set_index('ID')

df_groups=pd.read_csv('data/groups.csv', dtype = {'GroupYSID': str,'Quota': int,'Signed':int}).set_index('GroupYSID')

df_profs=pd.read_csv('data/profs-all.csv', dtype = {'ID': int, 'Name':str, 'Appointment':str,'SimpAppointment': 'category'}).set_index('ID')

df_teaching=pd.read_csv('data/teaching.csv', dtype = {'ProfID': int,'GroupYSID': str,'ProfType':'category', 'Days': str, 'Hours': str})

df_plans=pd.read_csv('data/plans_courses.csv', dtype={'CourseID':int,'PlanID':str,'CourseType':'category'})

df_plan_names=pd.read_csv('data/plans.csv')

# Hay algunas materias homólogas que no queremos que estén
df_homologas=pd.read_csv('data/homologas.csv')
df_homologas['old_name_code']=df_homologas['old_name_code'].apply(lambda x: '{:0>4d}'.format(x))
df_homologas['new_name_code']=df_homologas['new_name_code'].apply(lambda x: '{:0>4d}'.format(x))
```

```python
for df in [df_courses, df_groups, df_profs, df_teaching, df_plans]:
    display(df.sample(3))
```

## Agregar columnas y funciones útiles

Las columnas `GroupYSID` tienen información del grupo y semestre. Las colocamos por separado.

```python
df_teaching['Group']=df_teaching['GroupYSID'].apply(lambda x: x.split('-')[-2])
df_teaching['Semester']=df_teaching['GroupYSID'].apply(lambda x:x[:6])
df_teaching['Group'].head()
df_teaching['Semester'].head()
```

```python
def get_course_name(courseid):
    try:
        if type(courseid)==int:
            courseid='{:0>4d}'.format(courseid)
        if type(courseid)==str:
            return(df_courses.loc[courseid]['CourseName'])
    except:
        return "Not found"
    
def es_asignatura(text):
    if text=='PA':
        return(1)
    else:
        return(0)
```

## Join de la información

```python
df_teaching_join=df_teaching.join(df_groups,on='GroupYSID').join(df_profs, on='ProfID')
df_teaching_join['Asignatura']=df_teaching_join['SimpAppointment'].apply(es_asignatura)
df_teaching_join.head()

# Quitamos las materias homólogas. Dejamos sólo con el nombre reciente.
print(df_teaching_join.shape)
df_teaching_join=df_teaching_join[~df_teaching_join['Group'].isin(df_homologas['old_name_code'])]
print(df_teaching_join.shape)
```

### Antigüedad XFC

Para cada profesor, contamos en cuántos semestres regulares ha impartido clase dentro de los registros de XFC.

```python
sem_types=[['{}-1'.format(j) for j in range(2001,2022)],['{}-2'.format(j) for j in range(2001,2022)]]
sem_regular=sorted(sem_types[0]+sem_types[1])
```

```python
df_aux=df_teaching_join[(df_teaching_join['Asignatura']==1)&(df_teaching_join['Semester'].isin(sem_regular))&(df_teaching_join['ProfType']=='Profesor')&(df_teaching_join['Group'].isin(depa))]
df_aux['Semester2']=df_aux['Semester'].copy()
df_aux=df_aux.groupby(['ProfID']).agg({'Semester':'min','Semester2':'nunique','GroupYSID':'count'})
df_aux['AntigSem']=df_aux['Semester'].apply(lambda x: 42-sem_regular.index(x))
df_aux['AntigYea']=np.ceil(df_aux['AntigSem']/2)
df_aux['AvgCourse']=df_aux['GroupYSID']/df_aux['AntigSem']
df_aux=df_aux.sort_values(by='AvgCourse')
df_aux
```

```python
df_aux2=df_aux.groupby('AntigYea').agg({'AvgCourse':'mean'})
```

```python
plt.scatter(df_aux['AntigYea'],df_aux['AvgCourse'],alpha=0.1)
plt.scatter(df_aux2.index,df_aux2['AvgCourse'],color='green')
plt.xlabel('Antigüedad en años')
plt.ylabel('Promedio de cursos impartidos')
plt.title('Cursos impartidos según antigüedad')
plt.savefig("Figuras/promedio_por_antig.pdf")
```

```python
sem_types=[['{}-1'.format(j) for j in range(2001,2022)],['{}-2'.format(j) for j in range(2001,2022)]]
sem_regular=sorted(sem_types[0]+sem_types[1])
df_aux=df_teaching_join[(df_teaching_join['Asignatura']==0)&(df_teaching_join['Semester'].isin(sem_regular))&(df_teaching_join['ProfType']=='Profesor')&(df_teaching_join['Group'].isin(depa))]
df_aux['Semester2']=df_aux['Semester'].copy()
df_aux=df_aux.groupby(['ProfID']).agg({'Semester':'min','Semester2':'nunique','GroupYSID':'count'})
df_aux['AntigSem']=df_aux['Semester'].apply(lambda x: 42-sem_regular.index(x))
df_aux['AntigYea']=np.ceil(df_aux['AntigSem']/2)
df_aux['AvgCourse']=df_aux['GroupYSID']/df_aux['AntigSem']
df_aux=df_aux.sort_values(by='AvgCourse')
df_aux2=df_aux.groupby('AntigYea').agg({'AvgCourse':'mean'})
plt.scatter(df_aux['AntigYea'],df_aux['AvgCourse'],alpha=0.1)
plt.scatter(df_aux2.index,df_aux2['AvgCourse'],color='green')
plt.xlabel('Antigüedad en años')
plt.ylabel('Promedio de cursos impartidos')
plt.title('Cursos impartidos según antigüedad')
# plt.show()
plt.savefig("Figuras/promedio_por_antig_NA.pdf")
```
